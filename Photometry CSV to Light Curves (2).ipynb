```python
import pandas as pd
from astropy.io import fits
import numpy as np
import matplotlib.pyplot as plt
import photutils 
import math
import glob
import pdb
```

# Kepler 64 325

```python
k325 = pd.read_csv('kepler64-325.csv')
print(k325.head())
```

```python
#scale
fig = plt.figure(figsize=(10, 5))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])

#plot
plt.scatter(k325['Timestamp (JD)'], k325[' Kepler-64 : Magnitude (Centroid)'], label=" Kepler-64")
plt.scatter(k325['Timestamp (JD)'], k325[' UCAC4 650-079065 : Magnitude (Centroid)'], label='UCAC4 650-079065')

ax.plot(k325['Timestamp (JD)'], k325[' Kepler-64 : Magnitude (Centroid)'])
ax.plot(k325['Timestamp (JD)'], k325[' UCAC4 650-079065 : Magnitude (Centroid)'])

#legend
plt.legend()

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Magnitude (Centroid)')
plt.title('Light Curve for Kepler 64 on 11-20-2023')


# Show the plot
plt.show()
```
